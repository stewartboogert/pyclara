```python
import pybdsim
import matplotlib.pyplot as plt
```

```python
o = pybdsim.External.Pymad8.Output("../PMQ/TWISS_FEBE")
```

```python
o.getColumnsByKeys(['NAME','TYPE','L','K1','S']).iloc[0:20]
```

```python
p = pybdsim.External.Pymad8.Plot.Optics("../PMQ/TWISS_FEBE")

plt.subplot(3,1,1)
p.Beta()
plt.grid(True)
#plt.xlim(0,4)
#plt.ylim(0,100)
plt.axvline(2.354736,ls="--")

plt.subplot(3,1,2)
p.Alpha()
plt.grid(True)
#plt.xlim(0,4)
plt.axvline(2.354736,ls="--")

plt.subplot(3,1,3)
p.Mu()
plt.grid(True)
#plt.xlim(0,4)
plt.axvline(2.354736,ls="--")
```

### Make XSuite trackers before and after plasma

```python
xsuite_before = pyclara.Converters.elegant2xsuite("../PostInjector/FEBE.lte",
                                                  line_name="FEBE",
                                                  start_element="CLA-FEA-MAG-QUAD-13",
                                                  end_element="CLA-FEC1-SIM-FOCUS-01",
                                                  elegant_twi="../Run/FEBE.twi")

xsuite_after = pyclara.Converters.elegant2xsuite("../PostInjector/FEBE.lte",
                                                 line_name="FEBE",
                                                 start_element="CLA-FEC1-SIM-FOCUS-01",
                                                 end_element="CLA-FED-SIM-DUMP-01-START",
                                                 elegant_twi="../Run/FEBE.twi")

tracker_before = pyclara.Simulation.TrackerXSuite(xsuite_before)
tracker_after = pyclara.Simulation.TrackerXSuite(xsuite_after)

tracker = pyclara.Simulation.Beamline()
tracker.add_tracker(tracker_before)
tracker.add_tracker(tracker_after)
```
